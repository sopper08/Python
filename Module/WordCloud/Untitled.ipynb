```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt

d = path.dirname('./')
text = open(path.join(d, 'test.txt')).read()

wordcloud = WordCloud(max_font_size=60, prefer_horizontal=20, stopwords='school', background_color='white', mode='RGBA', normalize_plurals=False).generate(text)
# plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

wordcloud = WordCloud(max_font_size=60).generate(text)
# plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
